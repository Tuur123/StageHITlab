```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vectors = np.load('vectors.npy')
flow = vectors[0]
```

```python
flow.shape
```

```python
step = 3
fig = plt.figure(figsize=(16, 16), dpi=80)
fig.tight_layout()

plt.quiver(np.arange(0, flow.shape[1], step), np.arange(flow.shape[0], -1, -step), flow[::step, ::step, 0], flow[::step, ::step, 1])
```
